```python
import pandas as pd
import numpy as np

path = '/content/drive/MyDrive/Colab Notebooks/breast_cancer.csv'

#데이터 프레임 확인
df = pd.read_csv(path)

df
```

```python
#정답 데이터 제거 및 특정 데이터 생성
df=df.drop('label',axis=1)
df.head()
```

```python
#'worst compactness'를 예측 목표로 설정하고 특정 데이터 추출
y=df['worst compactness']
y.value_counts()
y
```

```python
#예측 목표 제외 및 특정 데이터 생성
X=df.drop('worst compactness',axis=1)
X.head()
```

```python
#학습 데이터와 테스트 데이터를 75:25로 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
# 선형 회귀 학습 및 세 가지 방식의 MSE(오차) 계산 비교
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))

# 실제 데이터와 모델 예측값의 분포 비교 시각화
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.scatter(y_test, y_test, label='Actual Values')
plt.scatter(y_test, ly_preds, c='y', label='Predicted Values')
plt.xlabel('Actual worst compactness')
plt.ylabel('Predicted worst compactness')
plt.title('Actual vs Predicted worst compactness')
plt.legend()
plt.show()
```

```python
# 결정 트리 회귀 모델 학습 및 MSE 오차 확인
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

model_dt = DecisionTreeRegressor(random_state=0)
model_dt.fit(X_train, y_train)

pred_dt = model_dt.predict(X_test)

print("--- Decision Tree Regressor ---")
print("평균제곱오차 (MSE):", mean_squared_error(y_test, pred_dt))
```

```python
# 랜덤 포레스트 회귀 모델 학습 및 MSE 오차 확인
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(n_estimators=100, random_state=0)
model_rf.fit(X_train, y_train)

pred_rf = model_rf.predict(X_test)

print("--- Random Forest Regressor ---")
print("평균제곱오차 (MSE):", mean_squared_error(y_test, pred_rf))
```
